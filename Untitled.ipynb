```python
import numpy as np
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
from datetime import date
day_month = list[31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
#df = distance_dataframe 
after_2000_df = df[df['term'] >= 2000] #limits 2000 or later
prd_df = after_2000_df[after_2000_df['term'] <= 2020] #limits to 2020 or earlier

prd_df['diff_days'] = prd_df['dateDecision'] - prd_df['dateArgument'] #gets differene between days
prd_df['diff_days'] = prd_df['diff_days']/np.timedelta64(1,'D') #expresses difference between days as integer

prd_df = prd_df[prd_df['diff_days'].notna()]

prd_df
```

```python
import matplotlib.pyplot as plt
dayslist = []

def meandiffdays(term):
    df = prd_df.loc[prd_df['term'] == term]
    return df['diff_days'].mean()
    
for i in np.arange(2000,2021):
    print("the mean diff days from term:", i, "is", meandiffdays(i), "days")
    dayslist.append(meandiffdays(i))
    
plt.plot(np.arange(2000,2021), dayslist)


    
    
    
```
